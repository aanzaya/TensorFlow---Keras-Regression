#### Feature Columns
    
* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the house’s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

### Separate all features from the labels

```python
X = df.drop('price',axis=1).values # For Tf you have to pass in the values
y = df['price'].values
```

For Tf you have to pass in the values

### TrainTest Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train, y=y_train,           
          validation_data=(X_test,y_test),
          batch_size=128, epochs=400)
```

Passing validation data on the model fit means after each epoch of training on the train data we'll quickly run the test data and check the loss function on the test data and this keep a tracking on how well performing on both train and test data. The test data will not affect the weight and biases of our networth. Keras will only use the train data to update the model and will not base it off the test data

The batch size is there since its a large dataset and hence we have to feed it to the model in batches. The smaller the batch size the longer training is going to take but the less likely you are going to overfit to the data since you are not feeding the whole dataset

## Model Valuation and Predictions

NB: If there Validation loss starts increasing at some point after the epochs, then it means we are overfitting our model
